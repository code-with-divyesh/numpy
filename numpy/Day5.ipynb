### Advance indexing in numpy

```python
import numpy as np 
x = np.array([[1, 2], [3, 4], [5, 6]]) 
y = x[[0,1,2], [0,1,0]] 
print(x)
print('The new array is :\n', y)
```

```python
a = np.arange(12).reshape(4,3)  # shape (4,3)
rows = [3,1,2]
print(a[rows])   # select those rows
```

```python
M = np.arange(20).reshape(5,4)
rows = [0,2,4]
print(M)
sub = M[rows]         
print(sub)
print(sub[sub % 3 == 0]) 
```

```python
arr = np.arange(6)
idx = [1,3,4]
print(arr[idx])
arr[idx] = -1
print(arr)
```

```python
a = np.arange(6)
v = a[1:4]   # view
print(v)
v[0] = 99
print(a)     # original changed

a = np.arange(6)
w = a[[1,2,3]]   # copy
w[0] = 88
print(a)         # unchanged
```
