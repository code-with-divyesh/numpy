Matrix Operations Project

Create two 3Ã—3 matrices and:

Add, subtract, multiply, and find dot product

Find determinant, transpose, and inverse
ðŸ‘‰ Concepts used: np.dot, np.linalg.det, np.transpose, np.linalg.inv

```python
import numpy as np
```

```python
arr1 = np.array([
    [2, 4, 6],
    [1, 3, 5],
    [7, 9, 11]
])

arr2 = np.array([
    [5, 8, 2],
    [6, 9, 3],
    [4, 7, 1]
])



print("addtion in two matrix ",arr1+arr2)
print("multiplication in two matrix:",arr1*arr2)
print("Dot product of two matrix:",np.dot(arr1,arr2))
print("transpose arr1:\n",np.transpose(arr1))
print("Determinant of arr1:",np.linalg.det(arr1))
print("inverse of arr2:",np.linalg.inv(arr2))
```
