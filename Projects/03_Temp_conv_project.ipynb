```python
import numpy as np

# Step 1: Create array of Celsius temperatures
celsius = np.array([0, 10, 25, 37, 100])
print("Celsius Temperatures:", celsius)

# Step 2: Convert to Fahrenheit (element-wise)
fahrenheit = (celsius * 9/5) + 32

# Step 3: Convert to Kelvin (element-wise)
kelvin = celsius + 273.15

# Step 4: Display results
print("\nFahrenheit Temperatures:", fahrenheit)
print("Kelvin Temperatures:", kelvin)

# Optional: Combine all in one 2D array for better view
combined = np.column_stack((celsius, fahrenheit, kelvin))
print("\nCombined Temperature Table (C, F, K):\n", combined)
```
