student Marks Analyzer

Input: Marks of 5 students in 5 subjects (2D array)

Task:

Calculate total marks and average of each student

Find topper

Find subject-wise highest and lowest marks
ðŸ‘‰ Concepts used: np.sum, np.mean, np.max, np.min, axis

```python
import numpy as np
```

```python
marks=np.array([
  [95,96,56,86,72],
  [96,89,55,63,74],
  [99,88,98,45,63],
  [96,86,85,99,63],
  [99,88,66,77,55]])
print(marks.shape)

total_marks = np.sum(marks, axis=1)
print("\nTotal Marks of each student:", total_marks)

average_marks=np.mean(marks,axis=1)
print("\n Average of each student:", average_marks)

topper_student = np.max(total_marks)
print("\nTopper student marks is:", topper_student)
topper_index = np.argmax(total_marks)
print("Topper is Student", topper_index + 1, "with marks", topper_student)


max_marks=np.max(marks,axis=0)
print("maximum marks per subject:",max_marks)

min_marks=np.min(marks,axis=0)
print("minimum marks per subject:",min_marks)
```
